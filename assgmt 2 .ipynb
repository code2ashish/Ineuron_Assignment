### Q1.Create the below pattern using nested for loop in Python.

```python
n= int(input("Enter a Number: "))

for i in range(n):
    for j in range(i+1):
        print("*",end="")
    print()
for i in range(n,1,-1):
    for j in range(i-1):
        print("*",end="")
    print()
```

### Q2. Write a Python program to reverse a word after accepting the input from the user.

```python
a=input("Input Word: ")
print("output: ",a[::-1])
```
