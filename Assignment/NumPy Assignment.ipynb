## Write a function so that the columns of the output matrix are powers of the input vector.

```python
def colpow(arg1,arg2):
    import numpy as np
    return np.array([[i**j for i in arg1 ] for j in range(1,arg2+1)] ).T
```

```python
colpow([1,2,8,7,5,6,2,5],3)
```

## Write a function to find moving average in an array over a window: Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

```python
def movavg(arg,k):
    li=[]
    for i in range(0,len(arg)-k+1):
        li.append((arg[i:k+i].sum())/k)
    
    print("Total Elements are: ",len(li))    
    return np.array(li)
```

```python
a=np.array( [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])

movavg(a,3)
```
